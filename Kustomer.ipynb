```python
import os
import pandas as pd

folder_path = os.getcwd()

# Get a list of files in the folder
files = os.listdir(folder_path)

#Iterate over each file
for filename in files:
    if filename.endswith('.csv') and 'Kustomer' not in filename:
        # Truncate the last 20 characters
        new_filename = filename[:-24] + filename[-4:]
        
        # Rename the file
        os.rename(os.path.join(folder_path, filename), os.path.join(folder_path, new_filename))

################################ import data ############################################################################################
############ After Hours Chats (found on Kustomer - Domo Dashboard
after_hours_chats_2024 = pd.read_csv('Kustomer2024-after-hours-chats.csv')
after_hours_chats_2025 = pd.read_csv('after-hours-chats.csv')
after_hours_chats = pd.concat([after_hours_chats_2024, after_hours_chats_2025])
after_hours_chats['Date'] = pd.to_datetime(after_hours_chats['DateTime'])
after_hours_chats['Source'] = 'After Hours Chats'

############  All Chats (found on Kustomer - Domo Dashboard
all_chats_2024 = pd.read_csv('Kustomer2024-all-chats.csv')
all_chats_2025 = pd.read_csv('all-chats.csv')
all_chats = pd.concat([all_chats_2024, all_chats_2025])
all_chats['Date'] = pd.to_datetime(all_chats['DateTime'])
all_chats['Source'] = 'All Chats'

############  Article Deflections
art_def_2024 = pd.read_csv('Kustomer2024-article-deflection-success-rate.csv')
art_def_2025 = pd.read_csv('article-deflection-success-rate.csv')
art_def = pd.concat([art_def_2024, art_def_2025])
art_def['Date'] = pd.to_datetime(art_def['DateTime'])
art_def['Source'] = 'Kustomer - Article Deflections'

############ Bot Attended Chats (found on Kustomer - Domo Dashboard
bot_attended_2024 = pd.read_csv('Kustomer2024-bot-attended.csv')
bot_attended_2025 = pd.read_csv('bot-attended.csv')
bot_attended = pd.concat([bot_attended_2024, bot_attended_2025])				 
bot_attended['Date'] = pd.to_datetime(bot_attended['DateTime'])
bot_attended['Source'] = 'Bot Attended Chats'

############  Conversations by User (found on Kustomer - Domo Dashboard 2
conv_by_user_2024 = pd.read_csv('Kustomer2024-conversations-by-user.csv')
conv_by_user_2025 = pd.read_csv('conversations-by-user.csv')
conv_by_user = pd.concat([conv_by_user_2024, conv_by_user_2025])					 
conv_by_user = pd.melt(conv_by_user, id_vars=['DateTime'], var_name='Agent', value_name='Conversation Channel Total' )
conv_by_user['Date'] = pd.to_datetime(conv_by_user['DateTime'])
conv_by_user['Source'] = 'Conversations by User'

############  Cnversations Marked Done (found on Kustomer - Domo Dashboard
# Unpivot Value Columns into one Conversation Channel Total column
conv_marked_done_2024 = pd.read_csv('Kustomer2024-conversations-marked-done-per-channel.csv')
conv_marked_done_2025 = pd.read_csv('conversations-marked-done-per-channel.csv')
conv_marked_done = pd.concat([conv_marked_done_2024, conv_marked_done_2025])				 
conv_marked_done = pd.melt(conv_marked_done, id_vars=['DateTime'], var_name='Conversation Channel', value_name='Conversation Channel Total') 
conv_marked_done['Date'] = pd.to_datetime(conv_marked_done['DateTime'])
conv_marked_done['Source'] = 'Conversations Marked Done'

############  Inbound Conversations (found on Kustomer - Domo Dashboard
# Unpivot Value Columns into one Conversation Channel Total column
inbound_2024 = pd.read_csv('Kustomer2024-inbound-conversations-by-day.csv')
inbound_2025 = pd.read_csv('inbound-conversations-by-day.csv')
inbound = pd.concat([inbound_2024, inbound_2025])
inbound = pd.melt(inbound, id_vars=['DateTime'], var_name='Conversation Channel',value_name='Conversation Channel Total')
inbound['Date'] = pd.to_datetime(inbound['DateTime'])
inbound['Source'] = 'Inbound Conversations by Day'

############  KIQ Conversations (found on Kustomer - Domo Dashboard
kiq_2024 = pd.read_csv('Kustomer2024-kiq-conversations.csv')
kiq_2025 = pd.read_csv('kiq-conversations.csv')
kiq = pd.concat([kiq_2024, kiq_2025])
kiq['Date'] = pd.to_datetime(kiq['DateTime'])
kiq['Source'] = 'KIQ'

############  live attended chats (found on Kustomer - Domo Dashboard
live_attended_2024 = pd.read_csv('Kustomer2024-live-attended.csv')
live_attended_2025 = pd.read_csv('live-attended.csv')
live_attended = pd.concat([live_attended_2024, live_attended_2025])					
live_attended['Date'] = pd.to_datetime(live_attended['DateTime'])
live_attended['Source'] = 'Live Attended'

############  Missed Chats Converted to Email (found on Kustomer - Domo Dashboard
missed_chats_2024 = pd.read_csv('Kustomer2024-missed-chats-converted-to-email.csv')
missed_chats_2025 = pd.read_csv('missed-chats-converted-to-email.csv')
missed_chats = pd.concat([missed_chats_2024, missed_chats_2025])					 
missed_chats['Date'] = pd.to_datetime(missed_chats['DateTime'])
missed_chats['Source'] = 'Missed Chats Converted to Email'

############  Avg Response Time for Chats (found on Kustomer - Domo Dashboard
response_time_chats_2024 = pd.read_csv('Kustomer2024-average-response-time-chats.csv')
response_time_chats_2025 = pd.read_csv('average-response-time-chats.csv')
response_time_chats = pd.concat([response_time_chats_2024, response_time_chats_2025])					 
response_time_chats['Date'] = pd.to_datetime(response_time_chats['DateTime'])
response_time_chats['Source'] = 'Avg. Response Time (Chats)'

############  Avg Response Time for Email (found on Kustomer - Domo Dashboard
response_time_email_2024 = pd.read_csv('Kustomer2024-average-response-time-email.csv')
response_time_email_2025 = pd.read_csv('average-response-time-email.csv')
response_time_email = pd.concat([response_time_email_2024, response_time_email_2025])
response_time_email['Date'] = pd.to_datetime(response_time_email['DateTime'])
response_time_email['Source'] = 'Avg. Response Time (Email)'

############  Avg Response Time for Phone (found on Kustomer - Domo Dashboard
response_time_phone_2024 = pd.read_csv('Kustomer2024-average-response-time-phone-seconds.csv')
response_time_phone_2025 = pd.read_csv('average-response-time-phone-seconds.csv')					 
response_time_phone = pd.concat([response_time_phone_2024, response_time_phone_2025])
response_time_phone['Date'] = pd.to_datetime(response_time_phone['DateTime'])
response_time_phone['Source'] = 'Avg. Response Time (Phone)'

############  Satisfaction (found on Kustomer - Domo Dashboard
satisfaction_2024 = pd.read_csv('Kustomer2024-overall-satisfaction-score.csv')
satisfaction_2025 = pd.read_csv('overall-satisfaction-score.csv')
satisfaction = pd.concat([satisfaction_2024, satisfaction_2025])
satisfaction = pd.melt(satisfaction,id_vars=['DateTime'],value_vars=['Positive Responses','Negative Responses'],var_name='Response Type',value_name='Response Count')
satisfaction['Date'] = pd.to_datetime(satisfaction['DateTime'])
satisfaction['Source'] = 'Satisfaction'

############  Support Contact Reasons (found on Kustomer - Domo Dashboard 2
support_2024 = pd.read_csv('Kustomer2024-support-contact-reasons.csv')
support_2025 = pd.read_csv('support-contact-reasons.csv')
support = pd.concat([support_2024, support_2025])					
support = pd.melt(support, id_vars=['DateTime'], var_name='Contact Reason', value_name='Count')
support['Date'] = pd.to_datetime(support['DateTime'])
support['Source'] = 'Support'

############  Taken vs. Missed Calls (found on Kustomer - Domo Dashboard
taken_vs_missed_calls_2024 = pd.read_csv('Kustomer2024-taken-vs-missed-calls-long-abandon.csv')
taken_vs_missed_calls_2025 = pd.read_csv('taken-vs-missed-calls-long-abandon.csv') 
taken_vs_missed_calls = pd.concat([taken_vs_missed_calls_2024, taken_vs_missed_calls_2025])
taken_vs_missed_calls['Date'] = pd.to_datetime(taken_vs_missed_calls['DateTime'])
missed_calls_may20_sep13 = pd.read_csv('Missed Calls - Not In Kustomer.csv')

# partition data before may 20 and after sep 13
taken_vs_missed_calls_before_may20 = taken_vs_missed_calls[taken_vs_missed_calls['Date'] <= '2024-05-19']
taken_vs_missed_calls_after_sep14 = taken_vs_missed_calls[taken_vs_missed_calls['Date'] >= '2024-09-14']

# partition data between may 20 - sep 13
taken_calls_may20_sep13 = taken_vs_missed_calls[
	(taken_vs_missed_calls['Date'] >= '2024-05-20') & 
	(taken_vs_missed_calls['Date'] <= '2024-09-13')]

# drop Missed Phone Calls column, set index to Date, change 'Assisted Phone Calls' to int
taken_calls_may20_sep13.drop(columns=['DateTime','Missed Phone Calls'],inplace=True)
taken_calls_may20_sep13['Date'] = taken_calls_may20_sep13['Date'].dt.strftime('%Y-%m-%d')
taken_calls_may20_sep13.set_index('Date',inplace=True)
taken_calls_may20_sep13['Assisted Phone Calls'] = taken_calls_may20_sep13['Assisted Phone Calls'].astype(int)

# convert Date to YYYY-MM-DD
missed_calls_may20_sep13['Date'] = pd.to_datetime(missed_calls_may20_sep13['Start Time'])
missed_calls_may20_sep13['Date'] = missed_calls_may20_sep13['Date'].dt.strftime('%Y-%m-%d')

# drop unnecessary columns
missed_calls_may20_sep13.drop(columns = ['Start Time','Agent','Result','Direction','Source','Consumer','Country','Queue','Dialed Number','Live Waiting Duration', 'Consumer Waiting Duration',
       'Talking Duration', 'Callback Waiting Duration',
       'Callback Pending Duration', 'Hold Duration', 'Wrap-up Duration',
       'Handling Duration', 'Ring Duration', 'Disposition', 'Flow Name',
       'Flow Duration', 'Flow Version', 'Waiting Type', 'Last Widget',
       'Required Skills/Proficiency', 'Optional Skills/Proficiency',
       'Agent Skills/Proficiency', 'Timezone'], inplace=True)

# group by Date to get daily Missed Phone Calls (rename Engagement ID to Missed Phone Callse
missed_calls_may20_sep13 = pd.DataFrame(missed_calls_may20_sep13.groupby('Date')['Engagement ID'].count())
missed_calls_may20_sep13.columns = ['Missed Phone Calls']
missed_calls_may20_sep13['Missed Phone Calls'] = missed_calls_may20_sep13['Missed Phone Calls'].astype(int)

# join missed calls may 20 - sep 13 onto taken calls
taken_vs_missed_calls_may20_sep13 = taken_calls_may20_sep13.join(missed_calls_may20_sep13)
taken_vs_missed_calls_may20_sep13['Missed Phone Calls'] = taken_vs_missed_calls_may20_sep13['Missed Phone Calls'].fillna(0)
taken_vs_missed_calls_may20_sep13['Date'] = taken_vs_missed_calls_may20_sep13.index

# concat back into one df
taken_vs_missed_calls = pd.concat([taken_vs_missed_calls_before_may20,taken_vs_missed_calls_may20_sep13,taken_vs_missed_calls_after_sep14])
taken_vs_missed_calls['Source'] = 'Taken vs. Missed Calls'

############  Transferred Chats (found on Kustomer - Domo Dashboard
transferred_2024 = pd.read_csv('Kustomer2024-transferred-chats.csv')
transferred_2025 = pd.read_csv('transferred-chats.csv')
transferred = pd.concat([transferred_2024, transferred_2025])
transferred['Date'] = pd.to_datetime(transferred['DateTime'])
transferred['Source'] = 'Transferred'

################################ Concat and save to csv for uploading to Domo via CSV Workbench ############################################################
Kustomer = pd.concat([
    after_hours_chats, all_chats, art_def, bot_attended, conv_by_user, conv_marked_done, inbound, 
	kiq, live_attended, missed_chats, response_time_chats, response_time_email, response_time_phone, 
    satisfaction, support, taken_vs_missed_calls, transferred
])

Kustomer.info()
Kustomer.to_csv('Kustomer - Domo Export.csv')
```
